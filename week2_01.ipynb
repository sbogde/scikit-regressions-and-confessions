```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame creation to simulate 'data'
data = pd.DataFrame({
    'Age': [23, 34, 45, 28, 34, 29, 23, 34, 42, 30, 28, 32, 40, 38, 35]
})

# Histogram for feature 'Age'
sns.histplot(data['Age'], bins=30, kde=True)  # KDE line smooths out the histogram
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
