```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Model with specific kernel and parameters
model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
```

```python
# Generate synthetic data for testing
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + 0.1 * np.random.randn(100)
```

```python
# Small ε vs Large ε
small_epsilon_model = SVR(kernel='rbf', epsilon=0.05)
large_epsilon_model = SVR(kernel='rbf', epsilon=0.5)
```

```python
# Fit models
small_epsilon_model.fit(X, y)
large_epsilon_model.fit(X, y)
```

```python
# Plot
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, small_epsilon_model.predict(X), color='red', label='Small ε=0.05')
plt.plot(X, large_epsilon_model.predict(X), color='green', label='Large ε=0.5')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Effect of Epsilon on SVR")
plt.legend()
plt.show()
```
