```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Model with specific kernel and parameters
model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
```

```python
# Generate synthetic data for testing
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + 0.1 * np.random.randn(100)
```

```python
# Low C vs High C
low_c_model = SVR(kernel='rbf', C=0.1, epsilon=0.1)
high_c_model = SVR(kernel='rbf', C=10, epsilon=0.1)
```

```python
# Fit models
low_c_model.fit(X, y)
high_c_model.fit(X, y)
```

```python
# Plot
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, low_c_model.predict(X), color='red', label='Low C=0.1')
plt.plot(X, high_c_model.predict(X), color='green', label='High C=10')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Effect of C on SVR")
plt.legend()
plt.show()
```
