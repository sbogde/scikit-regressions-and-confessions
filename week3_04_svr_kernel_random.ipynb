```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data for testing
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + 0.1 * np.random.randn(100)
```

```python
# 3 different kernels
linear_model = SVR(kernel='linear', C=1.0, epsilon=0.1)
poly_model = SVR(kernel='poly', degree=3, C=1.0, epsilon=0.1)
rbf_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)
```

```python
# Fit models
linear_model.fit(X, y)
poly_model.fit(X, y)
rbf_model.fit(X, y)
```

```python
# Plot
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, linear_model.predict(X), color='red', label='Linear Kernel')
plt.plot(X, poly_model.predict(X), color='green', label='Polynomial Kernel')
plt.plot(X, rbf_model.predict(X), color='purple', label='RBF Kernel')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Effect of Different Kernels in SVR")
plt.legend()
plt.show()
```
